```python
import math

print('\n ¡¡¡ BIENVENID@ !!! \n')
print('Qué operación desea realizar? \n')
print('1: Suma (+)')
print('2: Resta (-)')
print('3: Multiplicación (*)')
print('4: División (/)')
print('5: Potenciación (^)')
print('6: Radicación ()')
print('7: Logaritmación (log) \n')

opcion = input('Seleccione el número de la operación: ')

match opcion :
    case '1' :
        num1 = float(input('\nPor favor ingrese el 1er número: '))
        num2 = float(input('Por favor ingrese el 2do número: '))
        print('\nVamos a sumar (+) los valores ingresados:\n' )
        print(f'El resultado de la suma es: {(num1 + num2):,} \n')

    case '2' :
        num1 = float(input('\nPor favor ingrese el 1er número: '))
        num2 = float(input('Por favor ingrese el 2do número: '))
        print('\nVamos a restar (-) los valores ingresados:\n' )
        print(f'El resultado de la resta es: {(num1 - num2):,} \n')

    case '3' :
        num1 = float(input('\nPor favor ingrese el 1er número: '))
        num2 = float(input('Por favor ingrese el 2do número: '))
        print('\nVamos a multiplicar (*) los valores ingresados:\n' )
        print(f'El resultado de la multiplicación es: {(num1 * num2):,} \n')

    case '4' :
        num1 = float(input('\nPor favor ingrese el 1er número: '))
        num2 = float(input('Por favor ingrese el 2do número: '))
        if num2 == 0 :
            print('\n¡ No es posible la división entre cero (0) !\n')
            num1 = float(input('Por favor digite nuevamente el 1er número: '))
            num2 = float(input('Ahora digite el 2do número, que sea diferente de cero (0): '))
            print('\nVamos a dividir (*) los valores ingresados: \n' )
            print(f'El resultado de la división es: {(num1 / num2):,} \n')

    case '5' :
        num1 = float(input('\nPor favor ingrese la Base: '))
        num2 = float(input('Ahora ingrese el exponente: '))
        print(f'\nEl resultado de la Potencia es: {(num1 ** num2):,} \n')

    case '6' :
        num1 = float(input('\nPor favor ingrese el índice: '))
        if num1 == 0 :
            print('\n¡Error! No exite una raiz con índice cero (0) !\n')
            num1 = float(input('Por favor digite nuevamente un índice diferente de cero (o): '))
            num2 = float(input('\nAhora ingrese la cantidad subradical: '))
            print(f'\nLa raiz {int(num1)} de {num2} es: ',num2**(1/num1),'\n')
        else:
            num2 = float(input('\nAhora ingrese la cantidad subradical: '))
            print(f'\nLa raiz {int(num1)} de {num2} es: ',num2**(1/num1),'\n')

    case '7' :
        print('\nLog (base) argumento = logaritmo')
        num1 = float(input('\nPor favor ingrese la base : '))
        num2 = float(input('\nAhora ingrese el argumento: '))
        log_base_x = math.log(num2, num1)
        print(f'\nEL logaritmo en base {int(num1)} de {int(num2)} es igual a: {log_base_x} \n')

    case '_' :
        quit(print('Adios!'))

```
